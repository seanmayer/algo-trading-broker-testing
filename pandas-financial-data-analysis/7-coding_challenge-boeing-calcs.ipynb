# Coding Challenge: Boeing Calculations

```python
import pandas as pd
import yfinance as yf
import numpy as np
pd.options.display.float_format = '{:4f}'.format
```

```python
# Load Boeing (BA) data from 2014-10-01 to 2021-05-31
start = "2014-10-01"
end = "2021-05-31"
symbol = "BA"
ba = yf.download(symbol, start, end)
```

## 1. Calculate Boeing's Investment Multiple

```python
investment_multiple = ba['Close'].iloc[-1] / ba['Close'].iloc[0]
print(f"Boeing's Investment Multiple: {investment_multiple.item():.4f}")
```

## 2. Calculate Boeing's CAGR

```python
td = ba.index[-1] - ba.index[0]
td_years = td.days / 365.25
cagr = investment_multiple**(1/td_years) - 1
print(f"Boeing's CAGR: {cagr.item():.4%}")
```

## 3. Calculate Boeing's Geometric Mean Return

```python
ba['Returns'] = ba['Close'].pct_change()
n = ba['Returns'].count()
investment_multiple_from_returns = (1 + ba['Returns']).prod()
geometric_mean_return = investment_multiple_from_returns**(1/n) - 1
print(f"Boeing's Geometric Mean Return: {geometric_mean_return.item():.4%}")
```

## 4. Calculate Boeing's Investment Multiple with compound daily returns

```python
investment_multiple_compound = (1 + ba['Returns']).prod()
print(f"Boeing's Investment Multiple with compound daily returns: {investment_multiple_compound.item():.4f}")
```
