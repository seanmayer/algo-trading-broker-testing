### Summary
This notebook performs an analysis of Microsoft (MSFT) stock price changes, illustrating the difference between absolute and relative price changes.

**Columns Explained:**
 `MSFT`: The original closing price of MSFT stock for each date.`P_lag1`: The closing price of MSFT from the previous day (a one-period lag). This is use as a basis for calculating the change.`P_diff` / `P_diff2`: These columns represent the **absolute price change**, calculated a the simple difference between the current day's price and the previous day's price (`Price_Today - Price_Yesterday`). The notebook shows two different methods to achieve the same result. While straightforward, absolute changes can be misleading because a $1 change far more significant for a $10 stock than for a $500 stock. `Returns`: This column shows the **relative price change**, or the daily financial return

It is calculated as the percentage change from the previous day's closing price `(Price_Today - Price_Yesterday) / Price_Yesterday`. This metric is standardized and far more useful for financial analysis as it allows for the comparison of performance over time and across different assets, regardless of their absolute price levels.

**Why this is important:**
The primary goal of this notebook is to demonstrate why financial analysts and traders focus on **returns** rather than absolute price movements. Returns provide a normalized measure o
profitability and volatility, which is essential for building and evaluating trading strategies, conducting risk analysis, and comparing investment opportunities.
