# Short Selling Explainer

## What Is Short Selling?
Short selling is a strategy used when an investor expects a stock price to fall.

Instead of buying first and selling later, the short seller does the opposite:
1. Borrow shares from a broker
2. Sell those borrowed shares in the market
3. Buy shares back later to return to the lender

## How Profit and Loss Work
- If the price falls after the short sale, buying back later is cheaper, so the short seller profits.
- If the price rises, buying back is more expensive, so the short seller loses money.

This is the opposite payoff of a normal long position.

## Why Investors Short Stocks
Common reasons include:
- A valuation thesis (stock appears overpriced)
- A deteriorating business outlook
- Hedging another position in a portfolio
- Relative-value trades (long one asset, short another)

## Key Risks of Short Selling
Short selling has asymmetric risk:
- Maximum gain is limited (a stock cannot fall below zero).
- Potential loss is theoretically unlimited (a stock can keep rising).

Additional risks:
- Margin calls from the broker
- Borrow fees and stock-loan costs
- Forced buy-ins if borrow becomes unavailable
- Short squeezes (rapid price spikes driven by short covering)

## Practical Constraints
In real markets, shorting is constrained by:
- Share availability to borrow
- Broker risk controls and margin requirements
- Regulations that can temporarily restrict short sales

## Summary
Short selling is a way to express a bearish view or hedge risk by borrowing and selling shares first, then buying them back later. It can be useful, but it carries high risk and strict operational constraints, so position sizing and risk management are critical.
