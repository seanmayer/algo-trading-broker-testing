# Coding Challenge

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

## 1. Load Stock Price Data for GE and BTC from 2015-01-02 until 2020-12-31

```python
start = '2015-01-02'
end = '2020-12-31'
symbols = ['GE', 'BTC-USD']
df = yf.download(symbols, start, end)
```

## 2. Select Close prices only and create a price chart for GE

```python
ge_close = df['Close']['GE']
ge_close.plot(figsize=(15, 8), fontsize=13)
plt.legend(fontsize=13)
plt.title('GE Close Prices')
plt.show()
```

## 3. Normalise the stock prices for GE and BTC (Base Value 1) and visualise!

```python
close_prices = df['Close']
normalized_prices = close_prices.div(close_prices.iloc[0])
normalized_prices.plot(figsize=(15, 8), fontsize=13)
plt.legend(fontsize=13)
plt.title('Normalized Stock Prices')
plt.show()
```

## 4. What is the final normalised price for GE on 2020-12-30

```python
final_normalized_ge = normalized_prices['GE'].loc['2020-12-30']
print(f'The final normalized price for GE on 2020-12-30 was: {final_normalized_ge}')
```
