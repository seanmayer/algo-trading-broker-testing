Compound Returns and Geometric Mean Return

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.options.display.float_format = '{:4f}'.format
plt.style.use('seaborn-v0_8-darkgrid')
```

```python
msft = pd.read_csv("msft.csv", index_col = "Date", parse_dates = True)
msft
```

```python
multiple = (1+msft.Returns).prod() # alternative #3 to calc multiple (compund daily returns)
multiple
```

```python
multiple = (1+msft.Returns) # alternative #3 to calc multiple (compund daily returns)
multiple
```

```python
n = msft.Returns.count()
n
```

```python
geo_mean = multiple**(1/n) - 1 #Geometric mean return (daily)
geo_mean
```

```python
(1 + geo_mean)**n #alternative #4 to calculate multiple (geometric mean)
```

Compound returns, CAGR and geometric mean returns are closely related concepts

```python
mu = msft.Returns.mean() #athrimetic mean returns
mu
```

The arithmetic mean return is always greater than the geometric mean return and less useful

```python
(1+mu)**n # calculate multiple? not possible with arithmetic mean!
```
