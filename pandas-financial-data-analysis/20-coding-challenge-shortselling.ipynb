# Coding Challenge: Short Selling (Boeing)

```python
import pandas as pd
import numpy as np

pd.options.display.float_format = '{:,.6f}'.format
```

## 1) Calculate Daily Log Returns for Boeing

```python
close_prices = pd.read_csv('close.csv', index_col='Date', parse_dates=['Date'])

if 'BA' not in close_prices.columns:
    raise ValueError("Column 'BA' not found in close.csv")

boeing = close_prices[['BA']].dropna().rename(columns={'BA': 'Close'})
boeing['Log_Return_Long'] = np.log(boeing['Close'] / boeing['Close'].shift(1))
boeing = boeing.dropna(subset=['Log_Return_Long']).copy()

boeing.head()
```

## 2) Annualized Mean and Annualized Std for a Short Position

For log returns, short-position return is the negative of long-position return:

`Log_Return_Short = -Log_Return_Long`

Assume 252 trading days per year.

```python
trading_days_per_year = 252

boeing['Log_Return_Short'] = -boeing['Log_Return_Long']

daily_mean_short = boeing['Log_Return_Short'].mean()
daily_std_short = boeing['Log_Return_Short'].std(ddof=1)

annualized_mean_short = daily_mean_short * trading_days_per_year
annualized_std_short = daily_std_short * np.sqrt(trading_days_per_year)

print('Short Position (Boeing) Metrics:')
print(f'Daily mean log return (short):   {daily_mean_short:.8f}')
print(f'Daily std log return (short):    {daily_std_short:.8f}')
print(f'Annualized mean log return:      {annualized_mean_short:.8f}')
print(f'Annualized std (volatility):     {annualized_std_short:.8f}')
```

### Interpreting the Signs (Why One Negative and One Positive)

- `Log_Return_Short = -Log_Return_Long` for each day.
- So if Boeing's long-side average return is positive, the short-side average return is negative.

That is why the **annualized mean log return** for short BA is negative in this sample.

Volatility is different:
- Standard deviation measures spread (risk), not direction.
- It is always non-negative.
- Multiplying returns by `-1` flips the mean sign but leaves standard deviation unchanged.

So it is expected to see:
- negative mean return for the short position
- positive volatility for the same position

```python
summary = pd.DataFrame({
    'Metric': [
        'Annualized Mean Log Return (Short BA)',
        'Annualized Std / Volatility (Short BA)'
    ],
    'Value': [
        annualized_mean_short,
        annualized_std_short
    ]
})

summary
```
