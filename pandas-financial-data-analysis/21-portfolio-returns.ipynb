# Portfolio Returns: Simple vs Log Returns


This notebook demonstrates:
- how to calculate portfolio value
- how to calculate simple returns for individual assets and the portfolio
- why portfolio simple return equals the weighted average of constituent simple returns
- why this does not hold for log returns
- the key takeaway: log returns are time-additive, not asset-additive


```python
import pandas as pd
import numpy as np

pd.options.display.float_format = '{:,.6f}'.format
```

## 1) Create Asset Price Data


```python
# Simple two-asset example over one period (t0 -> t1)
prices = pd.DataFrame(
    {
        'Asset_A': [100, 112],
        'Asset_B': [100, 104],
    },
    index=['t0', 't1']
)

# Portfolio value is the sum of constituent asset values (1 unit each)
prices['Portfolio'] = prices['Asset_A'] + prices['Asset_B']

prices
```

## 2) Compute Simple Returns


```python
returns_simple = prices.pct_change()
returns_simple
```

```python
asset_a_simple = returns_simple.loc['t1', 'Asset_A']
asset_b_simple = returns_simple.loc['t1', 'Asset_B']
portfolio_simple = returns_simple.loc['t1', 'Portfolio']

print(f'Asset_A simple return:   {asset_a_simple:.2%}')
print(f'Asset_B simple return:   {asset_b_simple:.2%}')
print(f'Portfolio simple return: {portfolio_simple:.2%}')
```

## 3) Weighted Average of Simple Returns (50/50 Weights)


```python
weight_a = 0.5
weight_b = 0.5

weighted_simple = weight_a * asset_a_simple + weight_b * asset_b_simple

print(f'0.5 * Asset_A + 0.5 * Asset_B = {weighted_simple:.2%}')
print(f'Portfolio simple return         = {portfolio_simple:.2%}')
print(f'Match: {np.isclose(weighted_simple, portfolio_simple)}')
```

Using beginning-of-period weights, portfolio **simple return** equals the weighted average of constituent simple returns.


## 4) Compute Log Returns


```python
returns_log = np.log(prices / prices.shift(1))
returns_log
```

```python
asset_a_log = returns_log.loc['t1', 'Asset_A']
asset_b_log = returns_log.loc['t1', 'Asset_B']
portfolio_log = returns_log.loc['t1', 'Portfolio']
weighted_log = weight_a * asset_a_log + weight_b * asset_b_log

print(f'Asset_A log return:               {asset_a_log:.6f}')
print(f'Asset_B log return:               {asset_b_log:.6f}')
print(f'Portfolio log return (actual):    {portfolio_log:.6f}')
print(f'Weighted avg of asset log returns:{weighted_log:.6f}')
print(f'Difference (actual - weighted):   {portfolio_log - weighted_log:.10f}')
```

## 5) Interpretation

- Simple returns are **asset-additive** (with beginning-of-period weights).
- Log returns are **not asset-additive** across constituent assets.
- Log returns are **time-additive** for a single asset or portfolio path over multiple periods.
- Therefore, for cross-sectional portfolio aggregation, use simple returns.


## Key Takeaway

- Portfolio simple return = weighted average of constituent simple returns
- Portfolio log return != weighted average of constituent log returns
- Use simple returns for cross-sectional portfolio aggregation
- Use log returns when time additivity is the priority
