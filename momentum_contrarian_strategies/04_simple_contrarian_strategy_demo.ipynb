# Simple Contrarian Strategy Demo

> This notebook is for educational purposes only. It is a small made-up example designed to explain the logic of a contrarian trading strategy using `pandas` and `numpy`.

## Introduction

This notebook uses a small artificial price series to show the core mechanics of a simple contrarian strategy.

We will demonstrate:

1. how to calculate log returns from close prices
2. how to calculate a rolling mean of returns using a 3-period window
3. how to generate contrarian positions from that rolling mean
4. why the only difference between momentum and contrarian is the minus sign

The goal is to make the idea easy to follow before moving on to larger real-world datasets.

## Import Libraries

````python
import numpy as np
import pandas as pd
````

## Create a Small Example Dataset

We start with a simple made-up series of timestamps and close prices.

````python
# Create a small example DataFrame with timestamps and close prices.
data = pd.DataFrame({
    "timestamp": pd.date_range("2024-01-01 09:00:00", periods=8, freq="h"),
    "close": [1.1000, 1.1015, 1.1005, 1.1020, 1.1010, 1.1000, 1.1018, 1.1008]
})

# Set the timestamp column as the index to make the data look like a time series.
data.set_index("timestamp", inplace=True)

# Display the starting dataset.
data
````

## Calculate Log Returns

A log return compares the current close price with the previous close price.

Formula:

```python
np.log(current_close / previous_close)
```

The first row will be missing because there is no previous close price for the first observation.

````python
# Calculate log returns from the close price series.
data["returns"] = np.log(data["close"] / data["close"].shift(1))

# Display the DataFrame so we can see the new returns column.
data
````

## Calculate a 3-Period Rolling Mean of Returns

Next, we smooth the returns slightly by taking the rolling mean over the last 3 periods.

This tells us whether recent average returns have been positive or negative.

````python
# Calculate the rolling mean of returns using a 3-period window.
data["rolling_mean"] = data["returns"].rolling(window=3).mean()

# Display the intermediate result.
data
````

## Generate Momentum and Contrarian Positions

A momentum strategy follows recent movement:

```python
np.sign(rolling_mean)
```

A contrarian strategy does the opposite:

```python
-np.sign(rolling_mean)
```

This means:

- if the rolling mean return is positive, momentum goes `+1` and contrarian goes `-1`
- if the rolling mean return is negative, momentum goes `-1` and contrarian goes `+1`
- if the rolling mean return is zero, both become `0`

````python
# Generate momentum positions from the sign of the rolling mean return.
data["momentum_position"] = np.sign(data["rolling_mean"])

# Generate contrarian positions by flipping the momentum signal.
data["contrarian_position"] = -np.sign(data["rolling_mean"])

# Display only the columns we want to focus on.
data[[
    "close",
    "returns",
    "rolling_mean",
    "momentum_position",
    "contrarian_position"
]]
````

## Final Table

Below is the final beginner-friendly summary table.

````python
# Create a final view with the requested columns.
result = data[[
    "close",
    "returns",
    "rolling_mean",
    "momentum_position",
    "contrarian_position"
]].copy()

# Print the final DataFrame.
print(result)
````

## Key Takeaways

- Log returns are calculated from one close price to the next.
- A 3-period rolling mean gives a simple summary of recent return direction.
- Momentum uses `np.sign(rolling_mean)`.
- Contrarian uses `-np.sign(rolling_mean)`.
- The only coding difference between momentum and contrarian is the minus sign.
- This is a small educational example, but it shows the core idea clearly.
