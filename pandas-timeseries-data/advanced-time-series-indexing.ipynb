# Advanced Time Series Indexing

```python
# Import the pandas library for data manipulation
import pandas as pd
```

```python
# Load CSV file with datetime parsing and set datetime column as index
temp = pd.read_csv("data/temp.csv", parse_dates= ["datetime"], index_col= "datetime")
```

```python
# Access all data for February 20, 2015 using alternative date format
temp.loc["20FEBRUARY2015"]
```

```python
# Select multiple specific timestamps using a list
temp.loc[["2015-05-20 10:00:00", "2015-05-20 12:00:00"]]
```

```python
# Convert string dates to datetime index for reusable indexing
two_days = pd.to_datetime(["2015-05-20 10:00:00", "2015-05-20 12:00:00"])
# Display the datetime index
two_days
```

```python
# Use the datetime index variable to access data
temp.loc[two_days]
```

```python
# Create a daily resampler object (D = Daily)
temp.resample("D")
```

```python
# Convert resampler to list to see grouped data structure
list(temp.resample("D"))#[1][1]
```

```python
# Display first 25 rows of the dataset
temp.head(25)
```

```python
# Resample to daily frequency and sum all hourly values within each day
temp.resample("D").sum()
```

```python
# Resample to monthly frequency and calculate mean values (M = Monthly)
temp.resample("M").mean()
```

```python
# Resample to quarterly frequency and calculate mean values (Q = Quarterly)
temp.resample("Q").mean()
```

```python
# Resample to annual frequency and calculate mean values (A = Annual/Yearly)
temp.resample("A").mean()
```

```python
# Resample to 30-minute intervals and calculate mean (creates NaN for missing intervals)
temp.resample("30min").mean()
```

```python
# Resample to 15-minute intervals and calculate mean (upsampling with NaN)
temp.resample("15min").mean()
```

```python
# Resample to 10-minute intervals and calculate mean (higher frequency upsampling)
temp.resample("10min").mean()
```

```python
# Resample to 5-minute intervals and calculate mean (highest frequency upsampling)
temp.resample("5min").mean()
```
