```python
import pandas as pd
from ib_async import *
import time
util.startLoop()
ib = IB()
ib.connect()
```

```python
contract = Forex("EURUSD")
contract
```

```python
start = ib.reqHeadTimeStamp(contract, whatToShow = "MIDPOINT", useRTH = True)
start
```

```python
data = ib.reqHistoricalData(contract, "", barSizeSetting = "1 day", durationStr = "60 D",
                           whatToShow = "MIDPOINT", useRTH = True)
data
```

```python
df = util.df(data)
df
```

```python
durationStr: Time span of all bars: '60 S', '30 D', '13 W', '6 M', '10 Y'
```

```python
barSizeSettings: Time period of one bar, must be one of: '1 sec', '5 secs', '1 min', '2 mins', '3 mins', '5 mins', '10 mins', '15 mins', '20 mins', '30 mins', '1 hour', '2 hours', '3 hours', '4 hours', '8 hours', '1 day', '1 week', '1 month'
```

```python
df = util.df(ib.reqHistoricalData(contract, "", barSizeSetting = "5 mins", durationStr = "1 D",
                           whatToShow = "MIDPOINT", useRTH = True, formatDate = 2))
                           
df
```

```python
df = df.set_index("date").tz_convert("US/Eastern")
df
```
